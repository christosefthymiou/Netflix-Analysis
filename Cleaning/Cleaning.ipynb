```python
import pandas as pd
pd. get_option("display.max_columns")
Originals = pd.read_csv('../Data/Netflix Originals.csv', encoding='latin-1')
Titles = pd.read_csv('../Data/Netflix Titles.csv', encoding='latin-1')
```

```python
Originals.head(3)
```

```python
Originals['Title'] = Originals['Title'].str.replace(' ', '').replace('-', '').replace('&', '').replace(',', '').replace('!', '').replace('.', '').str.lower()
Originals.head(3)
```

```python
Originals.info()
```

```python
Titles.head(3)
```

```python
Titles['title'] = Titles['title'].str.replace(' ', '').replace('-', '').replace('&', '').replace(',', '').replace('!', '').replace('.', '').str.lower()
Titles.head(3)
```

```python
Titles.info()
```

```python
Data = Originals.merge(Titles, left_on = 'Title', right_on = 'title')
Data = Data[Data['country'].notna() & Data['rating'].notna() & (Data['type'] == 'Movie')]
Data = Data[['Title', 'Genre', 'Runtime', 'IMDB Score', 'Language', 'country', 'release_year', 'rating']]
Data.head(3)
```
