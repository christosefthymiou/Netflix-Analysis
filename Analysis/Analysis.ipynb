```python
x = Data.groupby(['Language']).count()['IMDB Score'].keys()
y = Data.groupby(['Language']).count()['IMDB Score'].values
porcent = 100.*y/y.sum()
patches, texts = plt.pie(y, startangle = 90, radius = 1.5)
labels = ['{0} - {1:1.2f} %'.format(i,j) for i, j in zip(x, porcent)]
sort_legend = True
if sort_legend:
    patches, labels, dummy =  zip(*sorted(zip(patches, labels, y), key=lambda x: x[2], reverse=True))
plt.legend(patches, labels, loc = 'left center', bbox_to_anchor = (-0.1, 1.), fontsize=8)
```
